# Q2:Linear classifier with update rule

```python
# training
def linear_classifier(weight,constant,lr,max_iter,data,label,x_train,y_train):
    random.shuffle(x_train.index.values)
    for i in x_train.index:
        ind = np.random.choice(x_train.index)
        feat = np.array(data.iloc[ind,])
        pred = np.sign(np.dot(feat,weight) + constant)
        if pred * label.iloc[i,] < 0:
            for j in range(max_iter):
                weight = weight + lr * feat * label[ind]
                constant = constant + label[ind]
                pred = np.sign(np.dot(feat,weight) + constant)
                if pred > 0:
                    break
    return weight,constant
```

```python
# testing crx data
for i in range(5):
    correct = 0
    weights,constant = linear_classifier(weights_crx,constant_crx,lr,max_iter,norm_crx,crx_label,crx_x_train,crx_y_train)
    for j in crx_x_test.index:
        feat = np.array(norm_crx.iloc[j,])
        pred = np.sign(np.dot(feat,weights) + constant)
        if pred == crx_label.iloc[j,]:
            correct += 1
    print("Linear Classifier Accuracy:{}".format(correct/len(crx_y_test)))
```

```python
# testing data
for i in range(5):
    correct = 0
    weights,constant = linear_classifier(weights_data,constant_data,lr,max_iter,norm_data,data_label,data_x_train,data_y_train)
    for j in range(len(data_x_test)):
        feat = np.array(norm_data.iloc[j,])
        pred = np.sign(np.dot(feat,weights) + constant)
        if pred == data_label.iloc[j,]:
            correct += 1
    print("Linear Classifier Accuracy:{}".format(correct/len(data_y_test)))
```

# Q3:Linear classifier with least-squared manner

Perhaps label encoder is inappropriate for these two data

```python
# least square method
def least_square(x,y):
    constant = np.ones((len(x),1))
    x = np.concatenate((np.matrix(x),constant),axis=1)
    xTx = np.matmul(x.transpose(),x)
    xTx_inv = np.linalg.inv(xTx)
    xTy = np.matmul(x.transpose(),y)
    weights = np.matmul(xTx_inv,xTy)
    return weights
```

```python
# testing crx
# since x_train is shuffle above,realign the index to follow y_train index
W = least_square(np.array(crx_x_train),crx_y_train)
constant_lse = np.ones((len(crx_x_test),1))
x = np.concatenate((constant_lse,np.matrix(crx_x_test)),axis=1)
pred = np.sign((np.dot(x,W.transpose())))
correct = 0
for i in range(len(pred)):
    if pred[i] == crx_label.iloc[i]:
        correct += 1
print("Least squared method Accuracy:{}".format(correct/len(pred))) 
```

```python
# testing data
# since x_train is shuffle above,realign the index to follow y_train index
W = least_square(np.array(data_x_train),data_y_train)
constant_lse = np.ones((len(data_x_test),1))
x = np.concatenate((constant_lse,np.matrix(data_x_test)),axis=1)
pred = np.sign((np.dot(x,W.transpose())))
correct = 0
for i in range(len(pred)):
    if pred[i] == data_label.iloc[i]:
        correct += 1
print("Least squared method Accuracy:{}".format(correct/len(pred)))  
```

# Q4:Voted perceptron

```python
def voted_perceptron(weight,constant,lr,data,label,x_train,y_train):
    random.shuffle(x_train.index.values)
    all_weight = []
    all_correct = []
    correct = 0
    for i in x_train.index:
        ind = np.random.choice(x_train.index)
        feat = np.array(data.iloc[ind,])
        pred = np.sign(np.dot(feat,weight) + constant)
        if pred * label.iloc[i,] > 0:
            correct += 1
        else:
            all_weight.append(weight)
            all_correct.append(correct)
            weight = weight + lr * feat * label[ind]
            constant = constant + label[ind]
            correct = 0
    return all_weight,constant,all_correct
```

```python
def avg_voted_perceptron(all_weight,all_correct):
    for i in range(len(all_weight)):
        all_weight[i] = all_weight[i]*all_correct[i]
    all_weight = sum(all_weight)/sum(all_correct)
    return all_weight
```

```python
# testing crx
for i in range(5):
    correct = 0
    all_weights,constant,all_correct = voted_perceptron(weights_crx,constant_crx,lr,norm_crx,crx_label,crx_x_train,crx_y_train)
    for j in crx_x_test.index:
        feat = np.array(norm_crx.iloc[j,])
        positive,negative = 0,0
        for k in range(len(all_weights)):
            pred = np.sign(np.dot(feat,all_weights[k]) + constant)
            if pred > 0:
                positive += pred * all_correct[k]
            else:
                negative += -pred * all_correct[k]
        if positive + negative > 0:
            pred = 1
        else:
            pred = -1
        if pred == crx_label.iloc[j,]:
            correct += 1
    print("Voted Perceptron Accuracy:{}".format(correct/len(crx_y_test)))
```

```python
# Average voted perceptron for crx
weights = avg_voted_perceptron(all_weights,all_correct)
correct = 0
for i in crx_x_test.index:
    feat = np.array(norm_crx.iloc[i,])
    pred = np.sign(np.dot(feat,weights) + constant)
    if pred == crx_label.iloc[i,]:
        correct += 1
print("Average voted perceptron Accuracy:{}".format(correct/len(crx_y_test)))
```

```python
# testing data
for i in range(5):
    correct = 0
    all_weights,constant,all_correct = voted_perceptron(weights_data,constant_data,lr,norm_data,data_label,data_x_train,data_y_train)
    for j in data_x_test.index:
        feat = np.array(norm_data.iloc[j,])
        positive,negative = 0,0
        for k in range(len(all_weights)):
            pred = np.sign(np.dot(feat,all_weights[k]) + constant)
            if pred > 0:
                positive += pred * all_correct[k]
            else:
                negative += -pred * all_correct[k]
        if positive + negative > 0:
            pred = 1
        else:
            pred = -1
        if pred == crx_label.iloc[j,]:
            correct += 1
    print("Voted Perceptron Accuracy:{}".format(correct/len(data_y_test)))
```

# Q5:Large margin classifier

```python
# compute distance
def distance(weights,observation,constant):
    dist = (np.dot(weights,observation)/np.linalg.norm(weights)) + constant
    return abs(dist)
```

```python
def max_margin_linear_classifier(weight,constant,lr,data,label,x_train,y_train):
    random.shuffle(x_train.index.values)
    sampled = []
    for i in x_train.index:
        ind = np.random.choice(x_train.index)
        sampled.append(ind)
        feat = np.array(data.iloc[ind,])
        pred = np.sign(np.dot(feat,weight) + constant)
        if pred * label.iloc[i,] <= 0:
            weight_new = weight + lr * feat * label[ind]
            constant_new = constant + label[ind]
            for j in sampled:
                all_dist,all_dist_new = 0,0
                all_dist = all_dist +  distance(weight,feat,constant)
                all_dist_new = all_dist_new + distance(weight_new,feat,constant_new)
                if all_dist_new < all_dist:
                    weight = weight_new
                    constant = constant_new
    return weight,constant
```

```python
# testing crx data
for i in range(5):
    correct = 0
    weights,constant = max_margin_linear_classifier(weights_crx,constant_crx,lr,norm_crx,crx_label,crx_x_train,crx_y_train)
    for j in crx_x_test.index:
        feat = np.array(norm_crx.iloc[j,])
        pred = np.sign(np.dot(feat,weights) + constant)
        if pred == crx_label.iloc[j,]:
            correct += 1
    print("linear classifier with the minimum w^2 Accuracy:{}".format(correct/len(crx_y_test)))
```

```python
# testing data
for i in range(5):
    correct = 0
    weights,constant = max_margin_linear_classifier(weights_data,constant_data,lr,norm_data,data_label,data_x_train,data_y_train)
    for j in data_x_test.index:
        feat = np.array(norm_data.iloc[j,])
        pred = np.sign(np.dot(feat,weights) + constant)
        if pred == data_label.iloc[j,]:
            correct += 1
    print("linear classifier with the minimum w^2 Accuracy:{}".format(correct/len(data_y_test)))
```

# Q6:Soft large margin classifier

```python
def max_margin_linear_classifier_with_slack(weight,constant,lr,data,label,x_train,y_train,slack):
    random.shuffle(x_train.index.values)
    sampled = []
    for i in x_train.index:
        ind = np.random.choice(x_train.index)
        sampled.append(ind)
        feat = np.array(data.iloc[ind,])
        pred = np.sign(np.dot(feat,weight) + constant)
        if pred * label.iloc[i,] <= 0:
            weight_new = weight + lr * feat * label[ind]
            constant_new = constant + label[ind]
            for j in sampled:
                all_dist,all_dist_new = 0,0
                all_dist = all_dist + distance(weight,feat,constant)
                all_dist_new = all_dist_new + distance(weight_new,feat,constant_new)
                if distance(weight,data.iloc[j,],constant) > slack:
                    all_dist = all_dist + distance(weight,feat,constant) * slack
                if distance(weight_new,data.iloc[j,],constant_new) > slack:
                    all_dist_new = all_dist_new + distance(weight_new,feat,constant_new) * slack
                if all_dist_new < all_dist:
                    weight = weight_new
                    constant = constant_new
    return weight,constant
```

```python
# testing crx data with constant = 1
slack = 1
for i in range(5):
    correct = 0
    weights,constant = max_margin_linear_classifier_with_slack(weights_crx,constant_crx,lr,norm_crx,crx_label,crx_x_train,crx_y_train,slack=slack)
    for j in crx_x_test.index:
        feat = np.array(norm_crx.iloc[j,])
        pred = np.sign(np.dot(feat,weights) + constant)
        if pred == crx_label.iloc[j,]:
            correct += 1
    print("linear classifier with the minimum w^2 and slack variable Accuracy:{}".format(correct/len(crx_y_test)))
```

```python
# testing crx data with different constant
slack_list = [1,3,5,7,9]
for i in slack_list:
    correct = 0
    weights,constant = max_margin_linear_classifier_with_slack(weights_crx,constant_crx,lr,norm_crx,crx_label,crx_x_train,crx_y_train,slack=i)
    for i in crx_x_test.index:
        feat = np.array(norm_crx.iloc[i,])
        pred = np.sign(np.dot(feat,weights) + constant)
        if pred == crx_label.iloc[i,]:
            correct += 1
    print("linear classifier with the minimum w^2 and slack variable Accuracy:{}".format(correct/len(crx_y_test)))
```

```python
# testing data with constant = 1
slack = 1
for i in range(5):
    correct = 0
    weights,constant = max_margin_linear_classifier_with_slack(weights_data,constant_data,lr,norm_data,data_label,data_x_train,data_y_train,slack=slack)
    for i in data_x_test.index:
        feat = np.array(norm_data.iloc[i,])
        pred = np.sign(np.dot(feat,weights) + constant)
        if pred == data_label.iloc[i,]:
            correct += 1
    print("linear classifier with the minimum w^2 and slack variable Accuracy:{}".format(correct/len(data_y_test)))
```
