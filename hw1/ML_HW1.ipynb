```python
# Normalization
def normalize(data):
    for i in data.columns:
        mean = data[i].mean()
        std = data[i].std()
        data[i] = (data[i] - mean)/std
    return data
```

# Q2:Linear classifier

```python
# initialize weights
weights = np.random.randn(len(crx.columns))
constant = np.random.randn(1)
lr = 0.001
weights,constant
```

```python
# compute the value
for i in x_train.index:
    ind = np.random.choice(x_train.index)
    val = np.array(crx.iloc[ind,])
    pred = np.sign(np.dot(val,weights) + constant)
    if pred <= 0:
        weights = weights + lr * val * crx_label[ind]
        constant += crx_label[ind]
# weights
```

```python
weights[np.where(weights < 0)] -= 1
weights
```

```python
# testing
correct = 0
for i in range(len(x_test)):
    val = np.array(crx.iloc[i,])
    pred = np.sign(np.dot(val,weights) + constant)
    if pred == y_test.iloc[i,]:
        correct += 1
print("Linear Classifier Accuracy:{}".format(correct/len(y_test)))
```

# Q3:Least square method


```python
# least square method
def least_square(x,y):
    constant = np.ones((len(x),1))
    x = np.concatenate((constant,np.matrix(x)),axis=1)
    y = y_train
    xTx = np.matmul(x.transpose(),x)
    xTx_inv = np.linalg.inv(xTx)
    xTy = np.matmul(x.transpose(),y)
    weights = np.matmul(xTx_inv,xTy)
    return weights
```

```python
W = least_square(x_train,y_train)
W
```

```python
# testing
constant = np.ones((len(x_test),1))
x = np.concatenate((constant,np.matrix(x_test)),axis=1)
pred = np.sign((np.dot(x,W.transpose())))
```

```python
correct = 0
for i in range(len(pred)):
    if pred[i] == y_test.iloc[i]:
        correct += 1
print("Least squared method Accuracy:{}".format(correct/len(pred)))  
```

# Q4:Voted perceptron

```python
all_weight = []
weights = np.random.randn(len(crx.columns))
constant = np.random.randn(1)
all_correct = []
correct = 0
```

```python
for i in x_train.index:
    ind = np.random.choice(x_train.index)
    val = np.array(crx.iloc[ind,])
    pred = np.sign(np.dot(val,weights) + constant)
    if pred >= 0:
        correct += 1
    else:
        all_weight.append(weights)
        all_correct.append(correct)
        weights = weights + lr * val * crx_label[ind]
        constant += crx_label[ind]
        correct = 0
```

```python
# testing
correct = 0
for i in x_test.index:
    val = np.array(crx.iloc[i,])
    positive,negative = 0,0
    for j in range(len(all_weight)):
        pred = np.sign(np.dot(val,all_weight[j]) + constant)
        if pred > 0:
            positive += pred * all_correct[j]
        else:
            negative += -pred * all_correct[j]
    if positive > negative:
        pred = 1
    else:
        pred = -1
    if pred == crx_label.iloc[i,]:
        correct += 1
print("Voted Perceptron Accuracy:{}".format(correct/len(y_test)))
```

```python
# Average voted perceptron
avg_vote = np.sum(all_weight,axis=0)/len(all_weight)
correct = 0
for i in x_test.index:
    val = np.array(crx.iloc[i,])
    pred = np.sign(np.dot(val,avg_vote) + constant)
    if pred == crx_label.iloc[i,]:
        correct += 1
print("Average voted perceptron Accuracy:{}".format(correct/len(y_test)))
```
